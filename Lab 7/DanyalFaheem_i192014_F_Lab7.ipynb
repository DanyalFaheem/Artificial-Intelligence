```python
from copy import deepcopy                        
import random                                  
import math                           
import numpy as np
board_size =0

def placeQueens(board):                          
    i = 0
    while (i < board_size):
        row = random.randint(0, board_size - 1)
        if (board[row][i] != "Q"):
            board[row][i] = "Q"
            i+=1

def getQueens(board):                            # This will get the positions of queens palced in the maze
    queen_positions = []                         
    for i in range(board_size):
        for j in range (board_size):
            if board[i][j] == "Q":
                temp = i,j
                queen_positions.append(temp)
    return queen_positions

def printBoard(board):                          
    for i in range(board_size):
        for j in range(board_size):
            print(board[i][j], end='  ')
        print()

def printPopulation(population):
    for i in range(len(population)):
        printBoard(population[i])
        print("\n")
    print("\n")

```

```python
def objective_function(board):  
    heuristic = 0  
    for i in board:
        heuristic += (i.count('Q') * (i.count('Q') - 1))
    
    queens = getQueens(board)
    diagonal = []
    for q in queens:
        i, j = q[0], q[1]
        y = j
        for x in range(i, board_size):
            if y < board_size:
                diagonal.append(deepcopy(board[x][y]))
            y += 1
        i, j = q[0], q[1]
        y = j - 1
        i = i - 1
        for x in range(i, -1, -1):
            if y > -1:
                diagonal.append(deepcopy(board[x][y]))
            y -= 1
        heuristic += (diagonal.count('Q') * (diagonal.count('Q') - 1))
        diagonal.clear()
    return heuristic        
    # This function will return the number of attacking queens
    # Step 1: Seeing if two queens are in same rows
    
    # Step 2 : Now we will check diagonals
    # There are a total of 4 diagonals
    
    # Top-Left Diagonal
    # Bottom-Right Diagonal
    # Bottom-Left Diagonal
    # Top-Right Diagonal
def expandBoard(board, queen_position):
    boards = []

    for i in range(len(queen_position)):
        x = deepcopy(queen_position[i][0])
        y = deepcopy(queen_position[i][1])   
        for j in range(board_size):
            if (x != j):
                b = deepcopy(board)
                b[x][y] = "*"
                b[j][y] = "Q"
                boards.append(b)
    return boards
def newBoard(board_size):
    rows , cols = (board_size,board_size)
    board = []                                       

    for i in range(rows):                         
        col = []
        for j in range(cols):                       
            col.append("*")                       
        board.append(col)
    return board
                
def createBoard(columns):
    board = newBoard(board_size)
    for i in range(board_size):
        board[columns[i]][i] = 'Q'
    
    return board

def makeChromosome(board):
    chromosome = []
    for j in range(board_size):
        for i in range(board_size):
            if board[i][j] == 'Q':
                chromosome.append(i)

    return chromosome


def createPopulation(size, board_size):
    population = []
    
    while len(population) < size:
        board = newBoard(board_size)
        placeQueens(board)
        if board not in population:
            population.append(board)
    
    return population

def sortPopulation(fringe, population):
    for i in range(len(fringe)):
        for j in range(i + 1, len(fringe) - 1):
            if fringe[j] < fringe[j + 1]:
                fringe[j], fringe[j + 1] = fringe[j + 1], fringe[j]
                population[j], population[j + 1] = population[j + 1], population[j]
    return fringe, population

def fitness_function(population):
    fitness = []
    for pop in population:
        fitness.append(objective_function(pop))

    return fitness


def parentSelection(population, fitness):
    fitness, population = sortPopulation(fitness, population)
    parents = []
    top = math.ceil((len(population) * 0.1))
    for i in range(top):
        parents.append(deepcopy(population[i]))

    return parents

def crossover(parents):
    generation = []
    point = random.randint(0, board_size - 1)
    for i in range(len(parents) - 1):
        parent1 = makeChromosome(parents[i])
        parent2 = makeChromosome(parents[i + 1])
        final = parent1[0:point] + parent2[point:]
        if final not in generation:
            generation.append(createBoard(final))
        final = parent2[0:point] + parent1[point:]
        if final not in generation:
            generation.append(createBoard(final))
        i += 1
    return generation

def mutation(generation):
    size = math.ceil(len(generation) / 2)
    change = set()
    for i in range(size):
        change.add(random.randint(0, len(generation) - 1))

    for c in change:
        chromosome = makeChromosome(generation[c])
        chromosome[random.randint(0, board_size - 1)] = random.randint(0, board_size - 1)
        generation[c] = deepcopy(createBoard(chromosome))

    return generation


def GeneticAlgorithm(generation_size, population):
    for i in range(generation_size):
        fitness = fitness_function(population)
        parents = parentSelection(population, fitness)
        generation = crossover(parents)
        mutated_generation = mutation(generation)
        for j in range(len(mutated_generation)):
            if objective_function(mutated_generation[j]) == 0:
                print("Solution found!")
                printBoard(mutated_generation[j])
        
        population = mutated_generation

    print("Iteration over!\nFinal mutated generation is")
    printPopulation(population)




def main():
    global board_size
    global population_size
    population_size = 100 
    board_size = 8
    population = createPopulation(population_size, board_size)
    GeneticAlgorithm(2, population)
    # printPopulation(population)
    
if __name__ == "__main__":

    main()
```

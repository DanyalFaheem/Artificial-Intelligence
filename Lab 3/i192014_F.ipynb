```python
def vacuumAgent():
    goalState = {"A": 0 , "B" : 0 , "C": 0}
    states = {"A": -1 , "B" : -1 , "C": -1}
    action = {0: "Clean", 1: "Dirty"}
    performance = 0
    currentLocation = input("Enter Location of Vacuum(A/B/C):").upper()
    copyLocation = currentLocation
    states[currentLocation] = int(input("Enter status of " + currentLocation + " (0/1)"))
    for state in states:
        if states[state] == -1:
            states[state] = int(input("Enter status of " + state + " (0/1)"))



    print("Goal State required:", goalState)
    while True:
        if states == goalState:
            break
        print("Vacuum is placed in", currentLocation)
        print("Location " + currentLocation + " is " + action[states[currentLocation]])
        if action[states[currentLocation]] == "Dirty":
            performance += 1
            print("Cost of Cleaning " + currentLocation + " is 1")
            print("Location " + currentLocation + " has been cleaned")
            states[currentLocation] = 0
        elif action[states[currentLocation]] == "Clean":
            print("Location " + currentLocation + " is clean")
        if getPreviousEntry(states, currentLocation) == "First":
            while getNextEntry(states, copyLocation) != "Last":
                if states[getNextEntry(states, copyLocation)] == 1:
                    currentLocation = getNextEntry(states, currentLocation)
                    performance += 1
                    print("Moving to location " + currentLocation + "\nCost of moving is 1")
                    isDirty = True
                else:
                    copyLocation = getNextEntry(states, copyLocation)
        else:
            isDirty = False
            copyLocation = currentLocation
            while getPreviousEntry(states, copyLocation) != "First":
                print(getPreviousEntry(states, copyLocation))
                if states[getPreviousEntry(states, copyLocation)] == 1:
                    currentLocation = getPreviousEntry(states, currentLocation)
                    performance += 1
                    print("Moving to location " + currentLocation + "\nCost of moving is 1")
                    isDirty = True
                else:
                    copyLocation = getPreviousEntry(states, copyLocation)
            if isDirty == False:
                while getNextEntry(states, copyLocation) != "Last":
                    print(getNextEntry(states, copyLocation))
                    if states[getNextEntry(states, copyLocation)] == 1:
                        currentLocation = getNextEntry(states, currentLocation)
                        performance += 1
                        print("Moving to location " + currentLocation + "\nCost of moving is 1")
                        isDirty = True
                    else:
                        copyLocation = getNextEntry(states, copyLocation)

                
        # if currentLocation == "A":
        #     if states["B"] == 1 or states["C"] == 1:
        #         performance += 1
        #         print("Moving to location B\nCost of moving is 1")
        #         currentLocation = "B"
        # elif currentLocation == "B":
        #     if states["A"] == 1:
        #         performance += 1
        #         print("Moving to location A\nCost of moving is 1")
        #         currentLocation = "A"
        #     elif states["C"] == 1:
        #         performance += 1
        #         print("Moving to location C\nCost of moving is 1")
        #         currentLocation = "C"
        # elif currentLocation == "C":
        #     if states["B"] == 1 or states["A"] == 1:
        #         performance += 1
        #         print("Moving to location B\nCost of moving is 1")
        #         currentLocation = "B"
    print("Performance Measurement", performance)
    print("Goal State required:", states)

def getNextEntry(dic, entry):
    counter = 0
    for d in dic:
        if counter == 1:
            return d
        if d == entry:
            counter = 1
    return "Last"

def getPreviousEntry(dic, entry):
    counter = 0
    for d in dic:        
        if d == entry:
            if counter == 0:
                return "First"
            else:
                return previousEntry
        counter += 1
        previousEntry = d

vacuumAgent()
# goalState = {"A": 0 , "B" : 0 , "C": 0}
# states = {"A": -1 , "B" : -1 , "C": -1}
# action = {0: "Clean", 1: "Dirty"}
# print(getNextEntry(states, "C"))
```
